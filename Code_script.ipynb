```python
import pandas as pd
```

```python
# Load the gene expression data and labels
data = pd.read_csv('/raw.csv')
labels = pd.read_csv('/labels.csv')

print('First 5 rows of data DataFrame:')
display(data.head())

print('\nFirst 5 rows of labels DataFrame:')
display(labels.head())
```

```python
# Drop any unnecessary ID columns if they exist (keep only the features)
# Ensure 'Class' is the column name for the cancer labels
if 'Unnamed: 0' in data.columns:
    data = data.drop(columns=['Unnamed: 0'])
if 'Unnamed: 0' in labels.columns:
    labels = labels.drop(columns=['Unnamed: 0'])

print('Data DataFrame after dropping "Unnamed: 0":')
display(data.head())

print('\nLabels DataFrame after dropping "Unnamed: 0":')
display(labels.head())
```

```python
# Merge them: Add the 'Class' column to the end of the gene expression dataset
data['Class'] = labels['Class']

print('Data DataFrame after merging Class column:')
display(data.head())
```

```python
# Save as a consolidated CSV for WEKA
data.to_csv('pancan_weka_ready.csv', index=False)
print("Data successfully prepared! Total shape:", data.shape)
```

```python
import pandas as pd

# 1. Load the master dataset headers (we only need the columns to save memory)
headers_df = pd.read_csv('pancan_weka_ready.csv', nrows=0)
all_genes = list(headers_df.columns)

# 2. Top indices extracted directly from your WEKA result files
# Python lists start at 0, while WEKA attributes start at 1, so we subtract 1.
weka_top_indices = [7965, 18382, 17110, 5408, 6594, 18747, 8350, 6817, 5578, 5577]
python_indices = [idx - 1 for idx in weka_top_indices]

# 3. Print out the mapped biological Gene Symbols
print("==============================================")
print("  IDENTIFIED TOP 10 PAN-CANCER BIOMARKERS     ")
print("==============================================")
for rank, py_idx in enumerate(python_indices, 1):
    gene_symbol = all_genes[py_idx]
    print(f"Rank {rank:02d} | WEKA Attribute: {weka_top_indices[rank-1]:<5} | Gene Symbol: {gene_symbol}")
print("==============================================")
```
